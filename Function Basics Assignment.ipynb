## Q10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

### Ans: 
### maps() : Applies function to all items in an iterable.

```python
numbers = [1, 2, 3, 4, 5]
squared_numbers = map(lambda x: x * x, numbers)

print(list(squared_numbers)) 
```

### reduce() : Reduces an iterable to a single value using a binary function.

```python
from functools import reduce

numbers = [1, 2, 3, 4]
product = reduce(lambda x, y: x * y, numbers)
print(product)
```

### filter() : Filters items in an iterable based on a function that returns a boolean

```python
numbers = [1, 2, 3, 4]
even = filter(lambda x: x % 2 == 0, numbers)
print(list(even))
```

## Q4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def check_prime(n):
    prime = []
    for num in range(2,n+1):
        is_prime = True
        for i in range(2,int(num**0.5) + 1):
            if num % i == 0:
                is_prime = False
                break
        if is_prime:
            prime.append(num)
    return prime
```

```python
check_prime(200)
```

## Q5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified 
## number of terms.

```python
class Fibonacci:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.current = 0
        self.next = 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        self.count += 1
        fib = self.current
        self.current, self.next = self.next, self.current + self.next
        return fib

fib_iter = Fibonacci(10)
for num in fib_iter:
    print(num)
```

## Q6.  Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def gen_power(n):
    for i in range(n+1):
        yield i**2

for i in gen_power(5):
    print(i)
```

## Q7.Implement a generator function that reads a file line by line and yields each line as a string. 

```python
def read_line(file_path):
    with open(file_path,'r') as file:
        for line in file:
            yield line.strip()
```

```python
for line in read_line("services.csv"):
    print(line)
```

## Q10.Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
def remove_vowels(s):
    return ''.join(filter(lambda x: x.lower() not in 'aeiou', s))
```

```python
remove_vowels("Hello Mike")
```

## Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the 
## product of the price per item and the quantity. The product should be increased by 10,- € if the value of the 
## order is smaller than 100,00 €.

## Write a Python program using lambda and map.

```python
orders = [
    [34587, "Learning Python", 4, 40.95],
    [98762, "Programming Python", 5, 56.80],
    [77226, "Head First Python", 3, 32.95],
    [88112, "Einführung in Python3", 3, 24.99]
]

def calculate_total(order):
    order_num, _, quantity, price = order
    total = quantity * price
    if total < 100:
        total += 10
    return order_num, total

totals = list(map(lambda order: calculate_total(order), orders))
print(totals)
```
